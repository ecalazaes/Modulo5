
## Gráfico de número de jogadores por país


```python
# Estilo do gráfico
plt.style.use('seaborn-v0_8-muted')

# Contagens
country_counts = df['Country'].value_counts()

# GRÁFICO: Jogadores por País
plt.figure(figsize=(12, 6))
bars1 = plt.bar(country_counts.index, country_counts.values, color='skyblue')
plt.title('Número de Jogadores por País')
plt.xlabel('País')
plt.ylabel('Número de Jogadores')
plt.xticks(rotation=90)

# Adicionar valores em cada barra (país)
for bar in bars1:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, f'{int(height)}',
             ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

## Gráfico de número de jogadores por posição

```python
# Estilo do gráfico
plt.style.use('seaborn-v0_8-muted')

# Contagens
position_counts = df['Position'].value_counts()

# GRÁFICO: Jogadores por Posição
plt.figure(figsize=(10, 6))
bars2 = plt.bar(position_counts.index, position_counts.values, color='salmon')
plt.title('Número de Jogadores por Posição')
plt.xlabel('Posição')
plt.ylabel('Número de Jogadores')
plt.xticks(rotation=0)

# Adicionar valores em cada barra (posição)
for bar in bars2:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, f'{int(height)}',
             ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

## Gráfico de distribuição da tecla para o Flash

```python
# Contagem geral das teclas de flash
flash_counts = df['FlashKeybind'].value_counts()

# Gráfico de pizza
plt.figure(figsize=(6, 6))
plt.pie(flash_counts, labels=flash_counts.index, autopct='%1.1f%%', colors=['#3498db', '#f39c12'])
plt.title('Distribuição do FlashKeybind (D ou F)')
plt.show()
```

##  Gráfico do uso da tecla para o Flash por posição

```python
# Criar tabela cruzada: número de jogadores por posição e tecla de flash
flash_by_position = pd.crosstab(df['Position'], df['FlashKeybind'])

# Converter para percentual dentro de cada posição
flash_percent = flash_by_position.div(flash_by_position.sum(axis=1), axis=0) * 100

# Plotar gráfico de barras empilhadas (percentual)
ax = flash_percent.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Set2')

# Adicionar os valores de percentual nas barras
for i, position in enumerate(flash_percent.index):
    bottom = 0
    for flash_key in flash_percent.columns:
        percent_value = flash_percent.loc[position, flash_key]
        if percent_value > 0:
            ax.text(
                i,
                bottom + percent_value / 2,  # posiciona o texto no centro da barra
                f'{percent_value:.1f}%',
                ha='center',
                va='center',
                fontsize=9,
                color='black'
            )
            bottom += percent_value

# Ajustes do gráfico
plt.title('Uso de FlashKeybind por Posição (%)')
plt.xlabel('Posição')
plt.ylabel('Porcentagem (%)')
plt.xticks(rotation=0)
plt.legend(title='Tecla Flash')
plt.tight_layout()
plt.show()
```

## Gráfico de Top 10 países com mais jogadores

```python
# Contar jogadores por país e pegar o top 10
top_countries = df['Country'].value_counts().head(10)

# Calcular porcentagens
total_players = df['Country'].value_counts().sum()
top_countries_percent = (top_countries / total_players) * 100

# Plotar gráfico de barras
plt.figure(figsize=(10, 6))
bars = plt.bar(top_countries.index, top_countries.values, color='cornflowerblue')

# Adicionar porcentagens em cima das barras
for bar, percent in zip(bars, top_countries_percent):
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + 1,
        f'{percent:.1f}%',
        ha='center',
        va='bottom',
        fontsize=9
    )

# Configurações do gráfico
plt.title('Top 10 Países com Mais Jogadores')
plt.xlabel('País')
plt.ylabel('Número de Jogadores')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Gráfico de média de kills por posição

```python
# Agrupar por posição e calcular média
media_kills = df.groupby('Position')['Avg kills'].mean().sort_values(ascending=False)

# Gráfico de barras
ax = media_kills.plot(kind='bar', color='orchid', title='Média de Kills por Posição no Campeonato')
plt.ylabel('Kills Médios')
plt.xlabel('Posição')
plt.xticks(rotation=0)
plt.tight_layout()

# Adicionar os valores nas barras
for i, v in enumerate(media_kills):
    ax.text(i, v + 0.05, f'{v:.2f}', ha='center', va='bottom')

plt.show()

```

## Gráfico de Top 10 jogadores com mais kills

```python
# Top 10 por kills
top_kills = df[['PlayerName', 'Avg kills']].sort_values(by='Avg kills', ascending=False).head(10)

# Gráfico de barras horizontais
ax = top_kills.plot(x='PlayerName', y='Avg kills', kind='barh', color='skyblue', legend=False)
plt.title('Top 10 Jogadores com Mais Kills')
plt.xlabel('Kills')
plt.ylabel('Jogador')
plt.gca().invert_yaxis()  # jogador com mais kills no topo
plt.tight_layout()

# Adicionar valores nas barras
for i, v in enumerate(top_kills['Avg kills']):
    ax.text(v + 0.1, i, f'{v:.2f}', va='center')

plt.show()

```
