# Manipulação de Arquivos

```python
texto = 'Cientista de Dados é a profissão mais legal do momento.\n'
texto = texto + 'O Senac tem um curso de Ciência de Dados.\n'
texto += 'Também existe uma pós-graduação em Big Data.'
```

```python
print(texto)
```

```python
# Importando o módulo os
import os
```

```python
# Criando um arquivo
arquivo = open(os.path.join('cientista.txt'),'w')
```

```python
# Gravando dados no arquivo
for palavra in texto.split():
  arquivo.write(palavra+' ')
```

```python
# Fechando o arquivo
arquivo.close()
```

```python
# Lendo o arquivo
arquivo = open('cientista.txt','r')
conteudo = arquivo.read()
arquivo.close()

print(conteudo)
#
```

```python
# Usando a expressão "with"
# Com o "with" o método close() é executado automaticamente
with open('cientista.txt','r') as arquivo:
  conteudo2 = arquivo.read()
```

```python
print(conteudo2)
```

```python
# Contando caracteres
print(len(conteudo2))
```

```python
print(texto)
```

```python
with open('cientista.txt','w') as arquivo:
  arquivo.write(texto[:21])
  arquivo.write('\n')
  arquivo.write(texto[:33])
```

```python
# Lendo o arquivo sem "with"
arquivo = open('cientista.txt','r')
conteudo = arquivo.read()
arquivo.close()

print(conteudo)
```

# Manipulaçao de arquivo CSV ( comma-separated values )

```python
# Importando o módulo csv
import csv
```

```python
with open('numeros.csv','w') as arquivo:
  writer = csv.writer(arquivo)
  writer.writerow(('primeira','segunda','terceira'))
  writer.writerow((55,93,76))
  writer.writerow((55,93,76))
```

```python
# Leitura de arquivos
with open('numeros.csv','r') as arquivo:
  leitor = csv.reader(arquivo)
  for x in leitor:
    print('Número de colunas:', len(x))
    print(x)
    print(type(x))
```

```python
# Gerando uma lista com dados do arquivo csv
with open('numeros.csv','r') as arquivo:
  leitor = csv.reader(arquivo)
  dados = list(leitor)

print(dados)
```

Manipulação JSON

```python
# Criando um dicionário
dict = {'nome': 'Erick Calazaes',
        'linguagem favorita': 'Java',
        'outras linguagens': ['Python','Nodejs','C'],
        'alunos': 1000}
#
```

```python
print(dict)
```

```python
for k,v in dict.items():
  print(k,v)
```

```python
print(dict.keys())
```

```python
print(dict.values())
```

```python
print(dict.items())
```

```python
print(dict['nome'])
```

```python
print(dict['linguagem favorita'])
```

```python
print(dict['outras linguagens'])
```

```python
# Importando módulo JSON
import json
```

```python
json.dumps(dict)
```

```python
# Criando um arquvio JSON
with open('dados.json','w') as arquivo:
  arquivo.write(json.dumps(dict))
```

```python
# Leitura de arquivo JSON
with open('dados.json','r') as arquivo:
  texto = arquivo.read()
  data = json.loads(texto)
```

```python
print(data)
```

```python
print(data['nome'])
```
