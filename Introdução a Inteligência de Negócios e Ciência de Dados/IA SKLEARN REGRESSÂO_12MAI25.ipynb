# Prevendo o preço de pizza da cidade do Rio de Janeiro com técnicas de regressão

```python
# importando o Matplotlib e Numpy
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# Suponha que você registrou o tamanho de pizzas que você comeu nos últimos meses

```python
# Diâmetros (cm)
diametros = [[7], [10], [15], [30], [45], [55]]

# Preços (R$)
precos = [[8], [12], [22], [38.5], [52], [65]]
```

```python
# Visualizando os dados com plot
plt.figure()
plt.xlabel('Diâmetro (cm)')
plt.ylabel('Preço (R$)')
plt.title('Pizza Rio de Janeiro Diâmetro x Preço')
plt.plot(diametros, precos, 'k.')
plt.axis([0, 60, 0, 70])
plt.grid(True)
plt.show()
```

```python
# Importando o módulo de Regressão Linear Scikit-Learn
from sklearn.linear_model import LinearRegression
```

```python
# Preparando os dados de treino

# Vamos Chamar os de X os dados de Diâmetro da pizza ( Dados que vc tem )
X = [[7], [10], [15], [30], [45], [55]]

# Vamos chamar de Y os dados de preço da pizza ( Dados que vc quer fazer a previsão )
Y = [[8], [12], [22], [38.5], [52], [65]]
```

# Y - É o valor previstao da variável dependente ( em nosso exemplo o preço da pizza )

# X - É a variável independente ou variável explanatória ( em nosso exemplo o diâmetro da pizza )

# Alfa: É o termo de interceptação ou coeficiente linear

# Beta: É o coeficiente de cada varíavel ou coeficiente angular.

```python
# Criando a varíavel do modelo
modelo = LinearRegression()
```

```python
# Treinando o modelo
modelo.fit(X, Y)
```

```python
# Vamos agora prever o preço da pizza de 35 cm de diâmetro
prev = modelo.predict([[35]])
print(prev)
```

```python
# Métricas de ML - Regressão Linear
# Coeficientes
print('Coeficiente: ', modelo.coef_)

# MSE ( mean squared error )
print('MSE: %.2f' % np.mean((modelo.predict(X) - Y) ** 2))

# Score de Variação: 1 é a predição perfeita
print('Score de Variação: %.2f' % modelo.score(X, Y))
```

```python
# Scatter plot representando a regressão linear
plt.scatter(X, Y, color='black')
plt.plot(X, modelo.predict(X), color='blue', linewidth=3)
plt.xlabel('X')
plt.ylabel('Y')
plt.xticks
plt.yticks
plt.show()
```
